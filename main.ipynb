```python



# Пример датафрейма (предполагается, что у вас уже есть df)
# df = pd.read_csv('titanic.csv')

# Для каждого сочетания Pclass и Sex вычисляем медианное или среднее значение для Age
# Можно использовать groupby и затем fillna()

# Для этого создадим функцию для заполнения пропусков
def fill_age(row, df_for_train):
    # Группируем по Pclass и Sex, и вычисляем медиану по Age
    age_median = df_for_train.groupby(['Pclass', 'Sex'])['Age'].median()
    # Если Age отсутствует, заполняем его медианой по соответствующему Pclass и Sex
    return age_median.loc[(row['Pclass'], row['Sex'], row['Embarked'], row['Fare'])] if pd.isna(row['Age']) else row['Age']

# Применяем функцию к каждой строке датафрейма
df_for_train['Age'] = df_for_train.apply(fill_age, axis=1, df_for_train=df_for_train)
df_for_train['Age'] = df_for_train.apply(fill_age, axis=1, df_for_train=df_for_train)
df_for_train['Age'] = df_for_train.apply(fill_age, axis=1, df_for_train=df_for_train)

# Или, если нужно заполнить средним значением, используем:
# df['Age'] = df.apply(lambda row: df.groupby(['Pclass', 'Sex'])['Age'].mean().loc[(row['Pclass'], row['Sex'])] if pd.isna(row['Age']) else row['Age'], axis=1)

# Проверим результат
df_for_train.head()
```

```python
# Создание диаграммы размаха
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_for_train, x='Pclass', y='Fare', palette="Set2", showfliers=True)
plt.title('скок стоит биллет')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()
```

```python
survival_rate = df_for_train.groupby(['Pclass', 'Sex'])['Survived'].mean().reset_index()

# Создание столбчатой диаграммы
plt.figure(figsize=(10, 6))  # Размер графика
sns.barplot(data=survival_rate, x='Pclass', y='Survived', hue='Sex', palette="Set2")

# Настройка графика
plt.title('%00 (of) Sex and Pclass ')
plt.xlabel('Pclass')
plt.ylabel('% (of) surviving')

# Отображение графика
plt.show()
```

```python
# Группировка данных по пункту отправления (Embarked) и вычисление процента выживших
survival_by_embarked = df_for_train.groupby(['Embarked', 'Survived']).size().unstack(fill_value=0)

# Построение круговой диаграммы для каждого пункта отправления
plt.figure(figsize=(12, 6))

for i, embarked in enumerate(survival_by_embarked.index):
    # Данные для каждого пункта отправления
    data = survival_by_embarked.loc[embarked]
    
    # Построение круговой диаграммы
    plt.subplot(1, 3, i+1)  # 1 строка, 3 столбца, индекс диаграммы
    data.plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999', '#66b3ff'], startangle=90, legend=False)
    
    # Настройка графика
    plt.title(f'Пункт отправления: {embarked}')
    plt.ylabel('')  # Убираем подпись оси Y

# Отображение графика
plt.tight_layout()
plt.show()
```

```python
quantitative_vars = df_for_train.select_dtypes(include=['float64', 'int64'])

# Вычисление корреляционной матрицы
corr_matrix = quantitative_vars.corr()

# Построение тепловой карты
plt.figure(figsize=(10, 8))  # Размер графика
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, cbar=True)

# Настройка графика
plt.title('Корреляционная матрица для количественных переменных')

# Отображение графика
plt.show()
```
